```python
from pyvi import ViTokenizer, ViPosTagger
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import MultinomialNB
import numpy as np
import gensim
import pandas as pd
```

# data preprocessing

```python
label = []
describe_train = []
data_train = []
data_test = []
describe_test = []
with open("trainning.txt",'r',encoding='utf-8') as f:
    lines_train = f.read()
    data_train = lines_train.split("\n")
f.close()
    

print(data_train[0])
```

```python
with open("testing.txt",'r',encoding='utf-8') as f_test:
    line_test = f_test.read()
    data_test = line_test.split("\n")
f_test.close()
```

```python
print(data_test[1])
```

```python
# train data
n_train = len(data_train)
for i in range(1,n_train - 1):
    name = data_train[i].split("\t")
    label.append(name[0])
    #name[1] = gensim.utils.simple_preprocess(name[1])
    name[1] = ViTokenizer.tokenize(name[1])
    describe_train.append(name[1])

# test data
n_test = len(data_test)
for i in range(0,n_test ):
    des = ViTokenizer.tokenize(data_test[i])
    describe_test.append(des)
```

```python
print(len(describe_test))
```

## ViTokenizer.tokenize(name[1]) dùng để tách từ trong văn bản

### sẽ chuyển thành => ông đã chấp_nhận lời mời tham_gia lễ kỷ_niệm năm ngày day của tổng_thống pháp jacque chirac  

# Tf-Idf Vectors as Features

```python
tfidf_vector = TfidfVectorizer(analyzer = 'word',max_features = 30000)
tfidf_vector.fit(describe_train)

result_train = tfidf_vector.transform(describe_train)
result_test = tfidf_vector.transform(describe_test)
print(result_train)
```

```python
# sau khi thực hiên TF-IDF, ma trận thu được có kích thước rất lướn -> khó trong viêcj tính toán
# => dùng thuật toán SVD để giảm số chiều của ma trận, vẫn giữ nguyên các thuộc tính ban đầu
svd = TruncatedSVD(n_components=300,random_state=42)
svd.fit(result_train)

result_train_svd = svd.transform(result_train)
result_test_svd = svd.transform(result_test)
```
